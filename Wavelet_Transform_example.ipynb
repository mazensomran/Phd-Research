```python
import cv2
from google.colab.patches import cv2_imshow
import pywt
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = "/content/drive/MyDrive/augmented_leather_def/Defcut3.png"
```

```python
input_img = cv2.imread(image)  # Read images
input_img = cv2.resize(input_img,(100, 100))
img = cv2.cvtColor(input_img, cv2.COLOR_BGR2GRAY)
# تحويل الموجات
coeffs = pywt.dwt2(img, 'haar')
cA, (cH, cV, cD) = coeffs

# عرض النتائج
cv2_imshow(input_img)
print('original')
cA = cv2.resize(cA,(100, 100))
cv2_imshow(cA)
print('approcsimation')
cH = cv2.resize(cH,(100, 100))
cv2_imshow(cH)
print('horisontal')
cV = cv2.resize(cV,(100, 100))
cv2_imshow(cV)
print('vertical')
cD = cv2.resize(cD,(100, 100))
cv2_imshow(cD)
print('diagonal')
```
